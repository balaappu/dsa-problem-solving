```python
def  print_x_pattern(word):
    n=len(word)
    for i in range(n):
        for j in range(n):
            if i==j or j ==(n-1-i):
                print(word[i],end='')
            else:
                print(' ',end='')    
        print()

word='sibiraj'
if len(word)%2==1:
    print_x_pattern(word)
else:
    print('please enter odd letter pattern')                    
```
