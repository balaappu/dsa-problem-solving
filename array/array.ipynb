```python
# #1)14. Sample Input â€“ Alternate Sorting 
# Input: {1, 2, 3, 4, 5, 6, 7}
# output: {7, 1, 6, 2, 5, 3, 4}
 
arr = [1, 2, 3, 4, 5, 6, 7]
n = len(arr)

arr.sort()  # Sort the array in ascending order

start = 0
end = n - 1
result = []

for i in range(n):
    if i % 2 == 0:
        result.append(arr[end])
        end -= 1
    else:
        result.append(arr[start])
        start += 1

print(" ".join(map(str, result)))
```

```python
# Given a string, change the order of words in the string (last string should come first).
# Should use RECURSION
  
# Sample:   one two three 
# Output :  three two one 


def reverse_words(s):

    words=s.split()
 # Base case: if only one word or empty
    if len(words)<=1:
        return s
      # Recursive case: reverse rest + first word  
    return reverse_words(' '.join(words[1:]))+' '+words[0]
s="one two three"
print(reverse_words(s))
```
