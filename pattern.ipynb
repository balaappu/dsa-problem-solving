```python
# X PATTERN 
# Print the word with odd letters â€“ PROGRAM
# Sample Output-+4477
# P     P
#  R   R
#   O O
#    G
#   R R
#  A   A
# M     M
def print_x_pattern(word):
    n = len(word)
    for i in range(n):
        for j in range(n):
            if i == j or j == (n - 1 - i):
                print(word[i], end='')
            else:
                print(' ', end='')
        print()

# Example usage:
word = "PROGRAM"
if len(word) % 2 == 1:
    print_x_pattern(word)
else:
    print("Please enter a word with odd length.")
```
