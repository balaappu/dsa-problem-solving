```python
def  print_x_pattern(word):
    n=len(word)
    for i in range(n):
        for j in range(n):
            if i==j or j ==(n-1-i):
                print(word[i],end='')
            else:
                print(' ',end='')    
        print()

word='sibiraj'
if len(word)%2==1:
    print_x_pattern(word)
else:
    print('please enter odd letter pattern')                    
```

```python
n=5
for i in range(1,n+1):
    for j in range(n,i,-1):
        print(" ",end=" ")
    for k in range(i,0,-1):
        print(k,end=" ")
    print()        
```
