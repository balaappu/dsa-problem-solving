```python
#PALINDROME
#  # Remove the palindrome
# Example 1:
# Input: Malayalam is my mother tongue
# Output: is my mother tongue
def remove_palindromes(sentence):
    words = sentence.split()
    filtered_words = [word for word in words if word.lower() != word.lower()[::-1]]
    return ' '.join(filtered_words)

# Example usage
input_str = "Malayalam is my mother tongue"
output_str = remove_palindromes(input_str)
print(output_str)
```

```python
def reverse_words(s):
    words = s.split()
    
    # Base case: if only one word or empty
    if len(words) <= 1:
        return s
    
    # Recursive case: reverse rest + first word
    return reverse_words(' '.join(words[1:])) + ' ' + words[0]
s = "one two three"
print(reverse_words(s))   # Output: three two one
```
